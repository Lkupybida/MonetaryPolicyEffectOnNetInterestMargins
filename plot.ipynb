```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_column(csv_file, column_name):
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Check if the column exists
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in the CSV file.")

    # Plot the chosen column against time
    plt.figure(figsize=(12, 6))
    plt.plot(df['Date'], df[column_name], marker='o')
    plt.title(f"{column_name} over Time")
    plt.xlabel("Date")
    plt.ylabel(column_name)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Display the plot
    plt.show()
```

```python
plot_column("data/extracted/complete/admin_expenses_differed.csv", "ПРИВАТБАНК")
```
